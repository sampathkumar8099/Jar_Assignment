### Data is clear there are no missing values or null values in data we can perform a head

```python
df["total_price"]=df["Unit price"]*df['Quantity']
```

```python
df.head()
```

### 1st Question :
Analyze the performance of sales and revenue at the city and branch level 

```python
Performance_of_Revenue_at_City_level = df.groupby("City").sum("total_price")[["total_price"]].reset_index()
```

```python
Performance_of_Revenue_at_City_level
```

```python
Performance_of_Revenue_at_Branch_level = df.groupby("Branch").sum("total_price")[["total_price"]].reset_index()
Performance_of_Revenue_at_Branch_level
```

```python
Performance_of_Sales_at_City_level = df.groupby("City").sum("Quantity")[["Quantity"]].reset_index()
Performance_of_Sales_at_City_level
```

```python
Performance_of_Sales_at_Branch_level
```

```python
Performance_of_Sales_at_Branch_level = df.groupby("Branch").sum("Quantity")[["Quantity"]].reset_index()
Performance_of_Sales_at_Branch_level
```

```python
Performance_of_Sales_at_Branch_level.to_excel("Performance_of_sales_at_Branch_level.xlsx")
Performance_of_Sales_at_City_level.to_excel("Performance_of_sales_at_City_level.xlsx")
Performance_of_Revenue_at_City_level.to_excel("Performance_of_Revenue_at_City_level.xlsx")
Performance_of_Revenue_at_Branch_level.to_excel("Performance_of_Revenue_at_Branch_level.xlsx")
```

### 2nd Question 
B. What is the average price of an item sold at each branch of the city 

```python
average_price_of_an_item_sold_at_each_branch_of_thecity = df.groupby(["Branch","City","Product line"]).mean("total_price")[["total_price"]].reset_index()
```

```python
average_price_of_an_item_sold_at_each_branch_of_thecity
```

```python
average_price_of_an_item_sold_at_each_branch_of_thecity.to_excel("average_price_of_an_item_sold_at_each_branch_of_thecity.xlsx")
```

### 3rd Task:
Analyze the performance of sales and revenue, Month over Month across the
Product line, Gender, and Payment Method, and identify the focus areas to
get better sales for April 2019

```python
 df['Date'] = pd.to_datetime(df['Date'])
Performance_of_Revenue_at_Every_month_by_Product_line = df.groupby([(pd.Grouper(key='Date', freq='M')),"Product line"]).sum("total_price")[["total_price"]].reset_index()
Performance_of_Revenue_at_Every_month_by_Product_line
```

```python
Performance_of_Sales_at_Every_month_by_Product_line = df.groupby([(pd.Grouper(key='Date', freq='M')),"Product line"]).sum("Quantity")[["Quantity"]].reset_index()
Performance_of_Sales_at_Every_month_by_Product_line
```

```python
Performance_of_Sales_at_Every_month_by_Gender = df.groupby([(pd.Grouper(key='Date', freq='M')),"Gender"]).sum("Quantity")[["Quantity"]].reset_index()
Performance_of_Sales_at_Every_month_by_Gender
```

```python
Performance_of_Revenue_at_Every_month_by_Gender = df.groupby([(pd.Grouper(key='Date', freq='M')),"Gender"]).sum("total_price")[["total_price"]].reset_index()
Performance_of_Revenue_at_Every_month_by_Gender
```

```python
Performance_of_Revenue_at_Every_month_by_Payment_method = df.groupby([(pd.Grouper(key='Date', freq='M')),"Payment"]).sum("total_price")[["total_price"]].reset_index()
Performance_of_Revenue_at_Every_month_by_Payment_method
```

```python
Performance_of_Sales_at_Every_month_by_Payment_method = df.groupby([(pd.Grouper(key='Date', freq='M')),"Payment"]).sum("Quantity")[["Quantity"]].reset_index()
Performance_of_Sales_at_Every_month_by_Payment_method
```

```python
Performance_of_Revenue_at_Every_month_by_Product_line.to_excel("Performance_of_Revenue_at_Every_month_by_Product_line.xlsx")
Performance_of_Sales_at_Every_month_by_Product_line.to_excel("Performance_of_Sales_at_Every_month_by_Product_line.xlsx")
Performance_of_Sales_at_Every_month_by_Gender.to_excel("Performance_of_Sales_at_Every_month_by_Gender.xlsx")
Performance_of_Revenue_at_Every_month_by_Gender.to_excel("Performance_of_Revenue_at_Every_month_by_Gender.xlsx")
Performance_of_Revenue_at_Every_month_by_Payment_method.to_excel("Performance_of_Revenue_at_Every_month_by_Payment_method.xlsx")
Performance_of_Sales_at_Every_month_by_Payment_method.to_excel("Performance_of_Sales_at_Every_month_by_Payment_method.xlsx")
```
